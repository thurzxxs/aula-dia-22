```python
#1 contagem simples, o usuario digita um numero e o programa imprimi do zero ate o numero

# entrada
numero = int(input("Digite um número para contar até ele: "))

for i in range(numero + 1):
    print(i)
```

```python
#2 contagem regressiva, o usuario digita um numero e o programa imprimi numero ate o zero

# entrada
number = int(input("Digite um número para iniciar a contagem regressiva: "))

for i in range(number, -1, -1):
    print(i)
```

```python
#5 soma o usuario, digita um numero e o programa soma os numeros do zero ate o numero

#entrada

numero= int(input("digite um numero:"))

total_soma = 0
for i in range(numero + 1):
    total_soma += i
print(f"A soma dos números de 0 a {numero} é: {total_soma}")
```

```python
#7 contar letras, o usuario digita uma palavra e o programa conta as letras das palavras

# entrada
palavra = input("Digite uma palavra: ")

# processamento e saída
numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
#8 contar vogais, o usuario digita uma palavra e o programa conta as vogais

# entrada
palavra = input("Digite uma palavra: ").lower()

# processamento
vogais = "aeiou"
contador_vogais = 0
for letra in palavra:
    if letra in vogais:
        contador_vogais += 1

# saída
print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
#9 maior numero digitado, o usuario digita quatro numeros e o programa imprimi o maior numero digitado

# entrada
print("Digite quatro números:")
numero1 = float(input("Número 1: "))
numero2 = float(input("Número 2: "))
numero3 = float(input("Número 3: "))
numero4 = float(input("Número 4: "))

# processamento
maior_numero = max(numero1, numero2, numero3, numero4)

# saída
print(f"O maior número digitado é: {maior_numero}")
```

```python
#10 o usuario digita um numero entre 0 e 100 e o programa mostra quais sao divisiveis por 3 5

# entrada
while True:
    try:
        numero_limite = int(input("Digite um número entre 0 e 100: "))
        if 0 <= numero_limite <= 100:
            break
        else:
            print("Por favor, digite um número entre 0 e 100.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

print(f"Números divisíveis por 3 e 5 até {numero_limite}:")

# processamento e saída
for i in range(numero_limite + 1):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```
